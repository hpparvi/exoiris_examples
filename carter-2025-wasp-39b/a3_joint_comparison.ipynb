This notebook compares the WASP-39b ExoIris results with the [Carter et al. (2024)](https://www.nature.com/articles/s41550-024-02292-x) results.

```python
import pandas as pd
from matplotlib.pyplot import subplots, setp
from exoiris import load_model, ExoIris
```

```python
spc1 = pd.read_csv('data/ZENODO/4_TRANSMISSION_SPECTRA/Fixed_LimbDarkening/NIRSpec_G395H_NRS1/bins_scale5.csv')
spc2 = pd.read_csv('data/ZENODO/4_TRANSMISSION_SPECTRA/Fixed_LimbDarkening/NIRSpec_G395H_NRS2/bins_scale5.csv')
```

```python
m_wn = load_model('WASP-39b-joint-k200.fits')
#m_gp = load_model('WASP-39b-NIRSpec-gp.fits')
sp_wn = m_wn.transmission_spectrum
#sp_gp = m_gp.transmission_spectrum
```

```python
from numpy import *
ts = m_wn
```

```python
ts.plot_setup(figsize=(13, 4));
```

## White noise analysis

```python
fig, ax = subplots(figsize=(12, 4))
ax.fill_between(sp_wn['wavelength'], 
                sp_wn['radius_ratio'] - sp_wn['radius_ratio_e'], 
                sp_wn['radius_ratio'] + sp_wn['radius_ratio_e'], 
                alpha=0.5, label='ExoIris WN', zorder=100)
ax.plot(sp_wn['wavelength'], sp_wn['radius_ratio'], 'k')

ax.errorbar(spc1.wave,
            spc1['rp/rs'], 
            [spc1['rp/rs_err_low'], spc1['rp/rs_err_hih']], 
            fmt='.', c='k', alpha=0.1, label='Carter et al. (2024)')
ax.errorbar(spc2.wave,
            spc2['rp/rs'], 
            [spc2['rp/rs_err_low'], spc2['rp/rs_err_hih']], 
            fmt='.', c='k', alpha=0.1, label=None)
#ax.plot(m_wn.k_knots, m_wn.posterior_samples.iloc[:, m_wn._tsa._sl_rratios].median().values, '.k')

ax.legend()
setp(ax, xlim=sp_wn['wavelength'].data[[20, -20]], ylim=(0.142, 0.153), 
     xlabel='Wavelength [$\mu$m]', ylabel='Radius ratio')
fig.tight_layout()
```
